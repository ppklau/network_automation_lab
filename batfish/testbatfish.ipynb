```python
# Connect to Batfish and init/select snapshot (run from notebook kernel)
import os
from pybatfish.client.session import Session
from pathlib import Path

BATFISH_HOST = os.environ.get("BATFISH_HOST", "192.168.1.240")
NETWORK_NAME = "acme_investments"
SNAPSHOT_NAME = "acme_lab"
# Resolve batfish directory robustly to avoid duplicated segments
# If the notebook kernel is launched from the repo root the batfish dir is './batfish'
# If the kernel cwd is already the 'batfish' directory, point to cwd instead.
cwd = Path.cwd()
if (cwd / 'snapshots').exists():
    BATFISH_DIR_ABS = cwd
elif (cwd / 'batfish' / 'snapshots').exists():
    BATFISH_DIR_ABS = cwd / 'batfish'
else:
    # fallback: prefer repo-relative 'batfish' directory if present
    candidate = (cwd / 'batfish').resolve()
    BATFISH_DIR_ABS = candidate if candidate.exists() else cwd
cur_path = os.path.dirname(__file__)
SNAPSHOT_PATH = "/Users/ppklau/stuff/ppklau/Development/NetworkAutomationLab/network_automation_lab/batfish/snapshots/acme_lab"

sess = Session(host=BATFISH_HOST)
sess.set_network(NETWORK_NAME)
sess.init_snapshot(str(SNAPSHOT_PATH), name=SNAPSHOT_NAME, overwrite=True)
sess.set_snapshot(SNAPSHOT_NAME)

# example queries
print("Snapshots:", sess.list_snapshots())
print("Nodes:", sess.q.nodeProperties().answer().frame().head())
```
